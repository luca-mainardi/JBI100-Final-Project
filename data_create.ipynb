```python
import pandas as pd
```

```python
def change_age(age_column_value):
    """
    Receives an age value with the format years-days
    :param age_column_value
    :return: The age in years only
    """

    # Split the string by '-' and take the first part
    age = age_column_value.split('-')[0]
    # Convert to int
    return int(age)


def country_to_continent(country_name):
    """
    Receives country name and converts it to continent name where it belongs to
    :param country_name
    :return: continent name
    """

    country_continent_mapping = {
        'Argentina': 'South America',
        'Australia': 'Oceania',
        'Belgium': 'Europe',
        'Brazil': 'South America',
        'Cameroon': 'Africa',
        'Canada': 'North America',
        'Costa Rica': 'North America',
        'Croatia': 'Europe',
        'Denmark': 'Europe',
        'Ecuador': 'South America',
        'England': 'Europe',
        'France': 'Europe',
        'Germany': 'Europe',
        'Ghana': 'Africa',
        'Iran': 'Asia',
        'IR Iran': 'Asia',
        'Japan': 'Asia',
        'Korea Republic': 'Asia',
        'Mexico': 'North America',
        'Morocco': 'Africa',
        'Netherlands': 'Europe',
        'Poland': 'Europe',
        'Portugal': 'Europe',
        'Qatar': 'Asia',
        'Saudi Arabia': 'Asia',
        'Senegal': 'Africa',
        'Serbia': 'Europe',
        'Spain': 'Europe',
        'Switzerland': 'Europe',
        'Tunisia': 'Africa',
        'United States': 'North America',
        'Uruguay': 'South America',
        'Wales': 'Europe'
    }

    country_continent_name = country_continent_mapping.get(country_name, 'Unknown')

    return country_continent_name
```

```python
player_keepers = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_keepers.csv")
player_keepersadv = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_keepersadv.csv")
player_defense = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_defense.csv")
player_stats = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_stats.csv")
player_misc = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_misc.csv")
player_gca = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_gca.csv")
player_passing = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_passing.csv")
player_shooting = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_shooting.csv")
player_possession = pd.read_csv("Data/FIFA World Cup 2022 Player Data/player_possession.csv")
player_defense.head()
```

```python
deleted_players = ['Ismail Mohamad','Kevin RodrÃ­guez','Yahya Jabrane']
```

```python
defenders_data = player_defense[player_defense['position'] == 'DF'][player_defense['minutes_90s'] > (15/90)][['player', 'age', 'minutes_90s',
                            'team', 'position', 'tackles','tackles_won', 'dribble_tackles_pct', 'blocked_passes', 'blocked_shots']]
defenders_data = defenders_data[~defenders_data['player'].isin(deleted_players)]
defenders_data = pd.merge(defenders_data, player_stats[['player', 'club']], on='player', how='left')
defenders_data = pd.merge(defenders_data, player_misc[['player', 'aerials_won_pct']], on='player', how='left')
defenders_data = pd.merge(defenders_data, player_passing[['player', 'passes_pct']], on='player', how='left')
defenders_data['tackles'] = defenders_data['tackles_won'] / defenders_data['tackles'].replace(0, 1)
defenders_data['blocked_passes'] = defenders_data['blocked_passes'] / defenders_data['minutes_90s'].replace(0, 1)
defenders_data['blocked_shots'] = defenders_data['blocked_shots'] / defenders_data['minutes_90s'].replace(0, 1)
defenders_data = defenders_data.drop(columns=['tackles', 'tackles_won'])    
defenders_data["age"] = defenders_data["age"].apply(change_age)
defenders_data["team_cont"] = defenders_data["team"].apply(country_to_continent)
defenders_data.to_csv('defenders_data.csv', index=False)
```

```python
midfielders_data = player_defense[player_defense['position'] == 'FW'][player_defense['minutes_90s'] > (15/90)][['player', 'age',
                                                         'minutes_90s', 'team', 'position', 'tackles', 'tackles_won', 'interceptions']]
midfielders_data = midfielders_data[~midfielders_data['player'].isin(deleted_players)]
midfielders_data = pd.merge(midfielders_data, player_stats[['player', 'club', 'goals_assists_per90']], on='player', how='left')
midfielders_data = pd.merge(midfielders_data, player_passing[['player', 'passes_pct']], on='player', how='left')
midfielders_data = pd.merge(midfielders_data, player_possession[['player', 'dribbles_completed_pct']], on='player', how='left')
midfielders_data = pd.merge(midfielders_data, player_gca[['player', 'sca_per90']], on='player', how='left')
midfielders_data['tackles'] = midfielders_data['tackles_won'] / midfielders_data['tackles'].replace(0, 1)
midfielders_data['interceptions'] = midfielders_data['interceptions'] / midfielders_data['minutes_90s'].replace(0, 1)
midfielders_data = midfielders_data.drop(columns=['tackles_won'])    
midfielders_data["age"] = midfielders_data["age"].apply(change_age)
midfielders_data["team_cont"] = midfielders_data["team"].apply(country_to_continent)
midfielders_data.to_csv('midfielders_data.csv', index=False)
```

```python
attackers_data = player_defense[player_defense['position'] == 'MF'][player_defense['minutes_90s'] > (15/90)][['player', 'age',
                                                                         'minutes_90s', 'team', 'position', 'tackles_interceptions']]
attackers_data = attackers_data[~attackers_data['player'].isin(deleted_players)]
attackers_data = pd.merge(attackers_data, player_stats[['player', 'club', 'goals_assists_per90']], on='player', how='left')
attackers_data = pd.merge(attackers_data, player_passing[['player', 'passes_pct']], on='player', how='left')
attackers_data = pd.merge(attackers_data, player_gca[['player', 'sca_per90']], on='player', how='left')
attackers_data = pd.merge(attackers_data, player_possession[['player', 'dribbles_completed_pct']], on='player', how='left')
attackers_data = pd.merge(attackers_data, player_shooting[['player', 'shots_on_target_pct']], on='player', how='left')
attackers_data['tackles_interceptions'] = attackers_data['tackles_interceptions'] / attackers_data['minutes_90s'].replace(0, 1)
attackers_data["age"] = attackers_data["age"].apply(change_age)
attackers_data["team_cont"] = attackers_data["team"].apply(country_to_continent)
attackers_data.to_csv('attackers_data.csv', index=False)
```

```python
keepers_data = player_keepers[player_keepers['minutes_90s'] > (15/90)][['player', 'team', 'age', 'club', 'position',
                                                                         'minutes_90s', 'gk_clean_sheets_pct','gk_save_pct', 'gk_pens_save_pct']]
keepers_data = keepers_data[~keepers_data['player'].isin(deleted_players)]
keepers_data = pd.merge(keepers_data, player_keepersadv[['player', 'gk_passes_pct_launched', 'gk_crosses_stopped_pct']], on='player', how='left')
keepers_data = pd.merge(keepers_data, player_passing[['player', 'passes_pct']], on='player', how='left')
keepers_data["age"] = keepers_data["age"].apply(change_age)
keepers_data["team_cont"] = keepers_data["team"].apply(country_to_continent)
keepers_data.to_csv('keepers_data.csv', index=False)
```
