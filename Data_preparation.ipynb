```python
# Load the data

# Player data
df_player_stats         = pd.read_csv('Data/FIFA World Cup 2022 Player Data/player_stats.csv', delimiter=',')
df_player_keepers       = pd.read_csv('Data/FIFA World Cup 2022 Player Data/player_keepers.csv', delimiter=',')
df_player_keepersadv    = pd.read_csv('Data/FIFA World Cup 2022 Player Data/player_keepersadv.csv', delimiter=',')
df_player_defense       = pd.read_csv('Data/FIFA World Cup 2022 Player Data/player_defense.csv', delimiter=',')
df_player_passing       = pd.read_csv('Data/FIFA World Cup 2022 Player Data/player_passing.csv', delimiter=',')
df_player_passing_types = pd.read_csv('Data/FIFA World Cup 2022 Player Data/player_passing_types.csv', delimiter=',')
df_player_shooting      = pd.read_csv('Data/FIFA World Cup 2022 Player Data/player_shooting.csv', delimiter=',')

# df_player_possession    = pd.read_csv('Data/FIFA World Cup 2022 Player Data/player_possession.csv', delimiter=',')
# df_player_gca           = pd.read_csv('Data/FIFA World Cup 2022 Player Data/player_gca.csv', delimiter=',')
# df_player_misc          = pd.read_csv('Data/FIFA World Cup 2022 Player Data/player_misc.csv', delimiter=',')
# df_player_playingtime   = pd.read_csv('Data/FIFA World Cup 2022 Player Data/player_playingtime.csv', delimiter=',')
```

```python
deleted_players = ['Ismail Mohamad','Kevin Rodr√≠guez','Yahya Jabrane']
```

```python
df_player_stats         = df_player_stats[~df_player_stats['player'].isin(deleted_players)]
df_player_keepers       = df_player_keepers[~df_player_keepers['player'].isin(deleted_players)]
df_player_keepersadv    = df_player_keepersadv[~df_player_keepersadv['player'].isin(deleted_players)]
df_player_defense       = df_player_defense[~df_player_defense['player'].isin(deleted_players)]
df_player_passing       = df_player_passing[~df_player_passing['player'].isin(deleted_players)]
df_player_passing_types = df_player_passing_types[~df_player_passing_types['player'].isin(deleted_players)]
df_player_shooting      = df_player_shooting[~df_player_shooting['player'].isin(deleted_players)]
```

```python
stats = df_player_stats.copy()
# keepers = pd.DataFrame()
defense = df_player_defense.copy()
# passing = pd.DataFrame()
shooting = df_player_shooting.copy()
```

```python
total_players = len(df_player_stats)
```

```python
stats['player_id'] = range(1, total_players + 1)
stats["age"] = stats["age"].apply(functions.change_age)
stats["team_cont"] = stats["team"].apply(functions.country_to_continent)
stats["goals_assists"] = stats["goals"] + stats["assists"]
stats = stats[['player_id','player','age','position','team','team_cont','club','games','games_starts','minutes','goals','assists','goals_assists','cards_yellow','cards_red']]
```

```python
stats
```

```python
keepers = pd.merge(df_player_keepers[['player','gk_goals_against','gk_shots_on_target_against','gk_saves','gk_clean_sheets','gk_pens_att','gk_pens_allowed','gk_pens_saved','gk_pens_missed']]
, df_player_keepersadv[['player','gk_free_kick_goals_against','gk_corner_kick_goals_against','gk_own_goals_against','gk_passes_launched','gk_passes_completed_launched','gk_passes','gk_passes_throws','gk_passes_length_avg','gk_goal_kicks','gk_goal_kick_length_avg','gk_crosses','gk_crosses_stopped']]
, on='player', how='outer')
keepers = pd.merge(keepers, stats[['player', 'player_id']], on='player', how='left')
id_col = keepers['player_id']
keepers.drop(labels=['player_id'], axis=1,inplace = True)
keepers.insert(0, 'player_id', id_col)
```

```python
defense['player_id'] = range(1, total_players + 1)
defense = defense[['player_id','player','tackles','tackles_won','tackles_def_3rd','tackles_mid_3rd','tackles_att_3rd','dribble_tackles','dribbled_past','blocks','blocked_shots','blocked_passes','interceptions','clearances','errors']]
defense = defense.astype({col: 'int64' for col in defense.select_dtypes(include=['float64']).columns})
```

```python
defense
```

```python
passing = pd.merge(df_player_passing[['player','passes','passes_completed','passes_short','passes_completed_short','passes_medium','passes_completed_medium','passes_long','passes_completed_long','assisted_shots','passes_into_final_third','passes_into_penalty_area','crosses_into_penalty_area','progressive_passes']]
, df_player_passing_types[['player','passes_live','passes_dead','passes_free_kicks','through_balls','passes_switches','crosses','throw_ins','corner_kicks','passes_offsides','passes_blocked']]
, on='player', how='outer')
passing = pd.merge(passing, stats[['player', 'player_id']], on='player', how='left')
id_col = passing['player_id']
passing.drop(labels=['player_id'], axis=1,inplace = True)
passing.insert(0, 'player_id', id_col)
passing = passing.astype({col: 'int64' for col in passing.select_dtypes(include=['float64']).columns})
```

```python
shooting['player_id'] = range(1, total_players + 1)
shooting = shooting[['player_id','player','shots','shots_on_target','shots_free_kicks','pens_att','pens_made',]]
shooting = shooting.astype({col: 'int64' for col in shooting.select_dtypes(include=['float64']).columns})
```

```python
goalkeepers = pd.merge(stats, keepers, on=['player_id', 'player'], how='inner')
goalkeepers = goalkeepers[goalkeepers.minutes>=50]
```

```python
dfs = [stats,defense,passing,shooting]
outfield_players = reduce(lambda left, right: pd.merge(left, right, on=['player_id', 'player'], how='inner'), dfs)
outfield_players = outfield_players[outfield_players['position']!='GK']
outfield_players = outfield_players[outfield_players.minutes>=50]
```

```python
# New columns
outfield_players["shots_off_target"] = outfield_players["shots"] - outfield_players["shots_on_target"]
outfield_players["passes_failed_short"] = outfield_players["passes_short"] - outfield_players["passes_completed_short"]
outfield_players["passes_failed_long"] = outfield_players["passes_long"] - outfield_players["passes_completed_long"]
outfield_players["tackles_failed"] = outfield_players["tackles"] - outfield_players["tackles_won"]
```
